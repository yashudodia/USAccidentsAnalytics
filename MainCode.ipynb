```python
import pandas as pd
import numpy as np
```

```python
i=pd.read_csv("./cleanedd_us_acc_data.csv")
pd.set_option('display.max_columns',None)
i
```

```python
i
```

```python
df_td = i[['Crossing','Stop','Traffic_Signal']]
print(df_td)
```

```python
traffic_count_dim = df_td[['Crossing','Stop','Traffic_Signal']].drop_duplicates().reset_index(drop=True)
traffic_count_dim['Traffic_id'] = traffic_count_dim.index
traffic_count_dim = traffic_count_dim[['Traffic_id','Crossing','Stop','Traffic_Signal']]
```

```python
traffic_count_dim
```

```python
df_timed = i[['Sunrise_Sunset','day_of_week','hour_of_day','month_of_year']]
print(df_timed)
```

```python
time_count_dim = df_timed[['Sunrise_Sunset','day_of_week','hour_of_day','month_of_year']].drop_duplicates().reset_index(drop=True)
time_count_dim['Time_id'] = time_count_dim.index
time_count_dim = time_count_dim[['Time_id','Sunrise_Sunset','day_of_week','hour_of_day','month_of_year']]
```

```python
time_count_dim
```

```python
location_count_dim = i[['Start_Lat','Start_Lng', 'Street', 'City', 'County', 'State', 'Zipcode',
       'Timezone', 'Airport_Code']].drop_duplicates().reset_index(drop=True)
location_count_dim['Location_id'] = location_count_dim.index
location_count_dim = location_count_dim[['Location_id','Start_Lat','Start_Lng','Street', 'City', 'County', 'State', 'Zipcode',
       'Timezone', 'Airport_Code']]
```

```python
location_count_dim
```

```python
weather_count_dim = i[['Weather_Timestamp','Weather_Condition','Precipitation(in)']].drop_duplicates().reset_index(drop=True)
weather_count_dim['Weather_id'] = weather_count_dim.index
weather_count_dim = weather_count_dim[['Weather_id','Weather_Timestamp','Weather_Condition','Precipitation(in)']]
```

```python
weather_count_dim
```

```python
i.columns
```

```python
i.head()
```

```python
fact_table=i.merge(traffic_count_dim,on=['Crossing','Stop','Traffic_Signal'])\
            .merge(time_count_dim,on=['Sunrise_Sunset','day_of_week','hour_of_day','month_of_year'])\
            .merge(location_count_dim,on=['Start_Lat','Start_Lng'])\
            .merge(weather_count_dim,on=['Weather_Timestamp','Weather_Condition','Precipitation(in)'])\
            [['ID', 'Severity', 'Start_Time', 'End_Time','Sunrise_Sunset','Crossing','Distance(mi)','Description','Time_id'
              ,'Weather_id','Traffic_id',
               'Weather_Timestamp', 'Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)',
               'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)','Precipitation(in)','Location_id', 'elapsed_time']]
```
